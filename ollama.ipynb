```python
import ollama
```

Creates the ollama server for python to access

```python
%%bash

# Start the server in the background
ollama serve > ~/ollserve.log 2>&1 &

# Wait a moment for the server to start
sleep 2
```

```python
resp = ollama.generate(
    model="llama3.1:8b",
    prompt="Define an experiment to measure the effect of caffeine on the ability for a student to focus",
    options={
        "temperature": 0.1, #Original setting: 0.1
        "num_ctx": 4096, # Original setting: 4096
        "top_p": 1.0, # Original setting: 1.0
        "num_predict": 10000,  # cap generation length
    },
)

print(resp["response"])
```

------------------------------------------------------------------------------------------------------------------------------------------
**FOR EXAMPLE 3**
ollama variables:
- temperature: 0-1-2,
    - increasing values increases creativity and volatility
- top_p: 0-1
    - increasing value increases variability in word choice
- num_ctx = context length: 100+
    - increases length of allowed input
- num_predict = max output length: 100+
    - increases length of allowed output
